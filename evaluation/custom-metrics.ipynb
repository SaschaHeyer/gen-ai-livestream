```python
import vertexai

PROJECT_ID = "sascha-playground-doit"
LOCATION = "us-central1"

EXPERIMENT_NAME = "quickstart-experiment-2"

vertexai.init(project=PROJECT_ID, location=LOCATION)

import pandas as pd
from vertexai.evaluation import EvalTask, PointwiseMetric, PointwiseMetricPromptTemplate
from vertexai.preview.evaluation import notebook_utils
from vertexai.generative_models import GenerativeModel
```

```python


# Define this custom "linguistic_acceptability" model-based metric.
linguistic_acceptability = PointwiseMetric(
    metric="linguistic_acceptability",
    metric_prompt_template=PointwiseMetricPromptTemplate(
        criteria={
            "Proper Grammar": "The language's grammar rules are correctly followed, including but not limited to sentence structures, verb tenses, subject-verb agreement, proper punctuation, and capitalization.",
            "Appropriate word choice": "Words chosen are appropriate and purposeful given their relative context and positioning in the text. Vocabulary demonstrates prompt understanding.",
        },
        rating_rubric={
            "5": "Excellent: The writing is grammatically correct, uses appropriate vocabulary",
            "4": "Good: The writing is generally grammatically correct, uses appropriate vocabulary",
            "3": "Satisfactory: The writing may have minor grammatical errors or use less-appropriate vocabulary",
            "2": "Unsatisfactory: The writing has significant grammatical errors, uses inappropriate vocabulary",
            "1": "Poor: The writing is riddled with grammatical errors, uses highly inappropriate vocabulary",
        },
    ),
)


print(linguistic_acceptability.metric_prompt_template)

responses = [
    "He can singing well, but he no dance.",
    "He can sing well, and he dances gracefully.",
]

eval_dataset = pd.DataFrame(
    {
        "response": responses,
    }
)
```

```python
eval_result = EvalTask(
    dataset=eval_dataset, metrics=[linguistic_acceptability], experiment=EXPERIMENT_NAME
).evaluate()
```
